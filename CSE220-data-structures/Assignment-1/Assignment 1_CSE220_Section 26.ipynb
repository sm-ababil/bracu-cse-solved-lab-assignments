```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence, beats):
    for i in beats:
        if i==1:
            temp = sequence[len(sequence)-1]
            for j in range(len(sequence)-1,0,-1):
                sequence[j]=sequence[j-1]
            sequence[0]=temp
    return sequence


# Test 02: Discard Cards
def removElement(arr, index):
    for i in range(index+1,len(arr)):
        arr[i-1]=arr[i]
    arr[len(arr)-1]=0

def discardCards(cards,num):
    i=0
    while i<len(cards):
        if cards[i]==num:
            removElement(cards, i)
        else:
            i+=1
    return cards


#Test 03: Merge Lineup
def mergeLineup(pokemon1, pokemon2):
    mergedArr = np.zeros(len(pokemon1), dtype = int)
    for i in range(len(pokemon1)):
        if pokemon1[i] == None and pokemon2[len(pokemon1)-1-i] == None:
            mergedArr[i] = 0
        elif pokemon1[i] == None:
            mergedArr[i] = 0+pokemon2[len(pokemon1)-1-i]
        elif pokemon2[len(pokemon1)-1-i] ==None:
            mergedArr[i] = pokemon1[i]+0
        else:
            mergedArr[i]= pokemon1[i]+pokemon2[len(pokemon1)-1-i]
    return mergedArr


# Test 04: Balance your Salami
def summ(arr):
    total = 0
    for i in arr:
        total+=i
    return total

def balanceSalami(salami):
    flag = False
    total = summ(salami)
    half = 0
    for i in range(len(salami)):
        half+=salami[i]
        if half == (total-half):
            flag = True
            break
    return flag
        

# Test 05: Protecc Salami
def protectSalami(salami):
    flag = False
    arr = [None]*len(salami)
    counted = [None]*len(salami)
    count = 0
    for i in range(len(salami)):
        if salami[i] not in counted:
            for j in range(i,len(salami)):
                if salami[j] == salami[i]:
                    count+=1
            if count!=1 and count in arr:
                flag = True
                break
            else:
                arr[i] = count
                counted[i] = salami[i]
                count = 0
    return flag


# Test 06: Odd Even Wave
def evenCount(arr):
    count = 0
    for i in arr:
        if i%2==0:
            count+=1
    return count

def oddEven(arr):
    for i in range(1,len(arr)-1):
        element = arr[i]
        if arr[i-1]%2 == 0 and arr[i]%2 == 0:
            for j in range(i+1, len(arr)):
                if arr[j]%2 !=0:
                    arr[i]=arr[j]
                    arr[j] = element
                    break
        elif arr[i-1]%2 != 0 and arr[i]%2 != 0:
            for j in range(i+1, len(arr)):
                if arr[j]%2 ==0:
                    arr[i]=arr[j]
                    arr[j] = element
                    break

def waveYourFlag(arr):
    if len(arr)%2 != 0:
        even = evenCount(arr)
        if even%2 != 0:
            if arr[0]%2 != 0:
                num = arr[0]
                for i in range(1, len(arr)):
                    if arr[i]%2 == 0:
                        arr[0] = arr[i]
                        arr[i] = num
                        break
        else:
            if arr[0]%2 == 0:
                num = arr[0]
                for i in range(1, len(arr)):
                    if arr[i]%2 != 0:
                        arr[0] = arr[i]
                        arr[i] = num
                        break
    oddEven(arr)
    return arr




```

```python
#Playing with Array and Probability Theory

def mean(arr):
    sum = 0
    for i in arr:
        sum+=i
    return sum/len(arr)

def standDev(arr):
    m = mean(arr)
    sqsum = 0
    for i in arr:
        sqsum += pow((i-m),2)
    return pow((sqsum/(len(arr)-1)),0.5)

def standAway(arr):
    m = mean(arr)
    s = standDev(arr)
    newArr = [None]*len(arr)
    count = 0
    for i in range(len(arr)):
        if (arr[i] <=(m - 1.5*s)) or (arr[i]>=(m + 1.5*s)):
            newArr[i] = arr[i]
            count+=1
            
    if count == 0:
        return None
    else:
        newArr2 = [None]*count
        flag = 0
        for j in newArr:
            if j != None:
                newArr2[flag] = j
                flag+=1

    return newArr2


        
arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(f"Mean: {mean(arr)}")
print(f"Standard Deviation: {standDev(arr)}")
print(f"Deviation away: {standAway(arr)}")

```
