```python
#Task 1

class Student:
    ID = 0

    def __init__(self, name, dept, age, cgpa):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa
        Student.ID +=1
    def showDetails(self):
        print(f"ID: {Student.ID}\nName: {self.name}\nDepartment: {self.dept}\nAge: {self.age}\nCGPA:{self.cgpa}")
    
    @classmethod
    def from_String(cls, info):
        lst = info.split("-")
        return Student(lst[0], lst[1], lst[2], lst[3])
        
s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()   
```

```python
#Task 2

class Assassin:

    total = 0

    def __init__(self, name, rate):
        if rate<=100:
            self.name = name
            self.rate = rate
            Assassin.total+=1
    def printDetails(self):
        print(f"Name: {self.name}\nSuccess rate: {self.rate}%\nTotal number of Assassin: {Assassin.total}")
    @classmethod
    def failureRate(cls, name, rate):
        return Assassin(name, 100 - rate)
    @classmethod
    def failurePercentage(cls, name, rate):
        return Assassin(name, 100 - int(rate[:-1]))

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#Task 7

class Cat:
    Number_of_cats = 0
    def __init__(self, color, work):
        self.color = color
        self.work = work
        Cat.Number_of_cats += 1
    @classmethod
    def no_parameter(cls):
        return Cat("White","sitting")
    @classmethod
    def first_parameter(cls, color):
        return Cat(color, "sitting")
    @classmethod
    def second_parameter(cls, work):
        return Cat("Grey", work)
    def changeColor(self,color):
        self.color = color
    def printCat(self):
        print(f"{self.color} cat is {self.work}")

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8

import math

class Cylinder:
    radius = 5
    height = 18

    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f"Updated: radius={self.radius} and height={self.height}")
    
    @classmethod
    def swap(cls, height, radius):
        return cls(radius, height)
    @classmethod
    def changeFormat(cls, value):
        radius, height = tuple(map(float, value.split("-")))
        return cls(radius, height)
    @staticmethod
    def area(radius, height):
        print(f"Area: {2*math.pi*radius**2 + 2*math.pi*radius*height}")
    @staticmethod
    def volume(radius, height):
        print(f"Volume: {math.pi*pow(radius,2)*height}")

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10

class SultansDine:
    total_branch = 0
    total_sell = 0
    dlist = []

    def __init__(self, branch):
        self.branch = branch
        SultansDine.total_branch+=1

    def sellQuantity(self, quantity):
        if quantity < 10:
            self.branch_sell = quantity * 300
        elif quantity < 20:
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400
        SultansDine.total_sell += self.branch_sell
        SultansDine.dlist.append(f"{self.branch}-{self.branch_sell}")
    def branchInformation(self):
        print(f"Branch Name: {self.branch}\nBranch Sell: {self.branch_sell} Taka")
    
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.total_branch}\nTotal Sell: {cls.total_sell} Taka")
        for item in cls.dlist:
            list1 = item.split("-")
            sell = int(list1[1])
            sell_rate = (sell/cls.total_sell)*100
            print(f"Branch Name: {list1[0]}, Branch Sell: {sell} Taka\nBranch consists of total sell's: {sell_rate:.2f}%")
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
