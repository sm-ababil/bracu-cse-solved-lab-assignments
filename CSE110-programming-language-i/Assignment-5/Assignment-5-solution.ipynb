```python
#Task 11

string1 = input("Please, enter the string: ")
string2 = string1.lower()
dict1={}
count=0

for i in string2:
    if i == " ":
        continue
    else:
        for j in string2:
            if i == j :
                count+=1
    dict1.update({i:count})
    count=0
print(dict1)
        
```

```python
#Task 13

list_1 = [("a", 1), ("b", 2), ("a", 3), ("b", 1), ("a", 2), ("c", 1)]
dict_1={}

for i in list_1:
	if i[0] in dict_1:
		continue
	else:
		list_2=[]
		for j in list_1:
			if i[0] in j:
				list_2.append(j[1])
	dict_1.update({i[0] : list_2})
print(dict_1)
```

```python
#Task 14

dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
    j = 0
    k = 22
    while j < 5:
        if j % 2 == 0:
            k = dict1[i] + j - (8 + k % 6) / 3
            dict1[i] = dict1[i]+ int(k)
        else:
            k = dict1[i] + j - (6 - k % 8) * 3
            dict1[i] = dict1[i] - int(k)
        j += 1
    print(int(k))
    print(i + " -> " + str(dict1[i]))

```
