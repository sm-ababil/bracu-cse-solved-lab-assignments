```python
#Task 4

def case(string1):
    uppercase = 0
    lowercase = 0
    for i in string1:  
        if 65<=ord(i)<=90:
            uppercase+=1
        elif 97<=ord(i)<=122:
            lowercase+=1

    print("No. of Uppercase characters :", uppercase)
    print("No. of Lowercase characters :", lowercase)

case("The quick Sand Man")
```

```python
#Task 7

def show_palindrome(x):
    palindrome=""
    for i in range(1,x+1):
        palindrome+=str(i)
    for j in range(x-1,0,-1):
        palindrome+=str(j)
    return palindrome

print(show_palindrome(5))
```

```python
#Task 11

def rem_duplicate(tup):
    list1 = list(tup)
    list2 = []
    for i in list1:
        if i not in list2:
            list2.append(i)
    return tuple(list2)

print(rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2])))
```

```python
#Task 12

def max2(list1):
    list_name=[]
    removed=0
    for i in list1:
        if list_name.count(i)<2:
            list_name.append(i)
        else:
            removed+=1
    print("Removed:",removed)
    return list_name

print(max2([10, 10, 15, 15, 20]))

```

```python
#Task 13

operator = input("Enter the operator: ")
operand1 = float(input("Enter the first number: "))
operand2 = float(input("Enter the second number:"))

def calculation(operator, operand1, operand2):
    solution = 0
    if operator == "+":
        solution = operand1 + operand2
    elif operator == "-":
        solution = operand1 - operand2
    elif operator == "*":
        solution = operand1 * operand2
    elif operator == "/":
        solution = operand1 / operand2

    return solution

print(calculation(operator, operand1, operand2))
```

```python
#Task 14

sentence = input("Please, enter trhe sentence: ")
position = int(input("Please, enter the position: "))

def rem_index(sentence, position):
    new = ""
    removed = ""
    for i in range(len(sentence)):
        if i != 0 and i % position == 0:
            removed+=sentence[i]
        else:
            new+=sentence[i]
    new_sentence = new+removed
    return new_sentence
        
print(rem_index(sentence, position))
        

```

```python
#Task 15

def price(item_list, location="Dhanmandi"):
    dict1={"Rice":105, "Potato":20, "Chicken":250, "Beef":510, "Oil":85}
    total=0
    for i in item_list:
        if i not in dict1:
            continue
        else:
            total+=dict1[i]
    if location != "Dhanmandi":
        total+=70
    else:
        total+=30
    return total

item_list=[]
quantity = int(input("Enter the number of item: "))
for i in range(quantity):
    item=input("Enter the item name: ")
    item_list.append(item)

print(price(item_list, "Mohakhali"))
```
