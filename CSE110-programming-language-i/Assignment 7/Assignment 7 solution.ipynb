```python
#Sorting

def selection_asc(mylist):
    for idx in range(len(mylist)-1):
        min_val = mylist[idx]
        min_idx = idx
        for idx2 in range(idx+1, len(mylist)):
            if mylist[idx2]<min_val:
                min_val = mylist[idx2]
                min_idx = idx2
        temp = min_val
        mylist[min_idx] = mylist[idx]
        mylist[idx] = temp
    return mylist

def selection_dsc(mylist):
    for idx in range(len(mylist)-1):
        max_val = mylist[idx]
        max_idx = idx
        for idx2 in range(idx+1, len(mylist)):
            if mylist[idx2]>max_val:
                max_val = mylist[idx2]
                max_idx = idx2
        temp = max_val
        mylist[max_idx] = mylist[idx]
        mylist[idx] = temp
    return mylist

def bubble_asc(mylist):
    for idx in range(len(mylist)-1):
        for idx1 in range(len(mylist)-idx-1):
            if mylist[idx1]>mylist[idx1+1]:
                temp=mylist[idx1]
                mylist[idx1]=mylist[idx1+1]
                mylist[idx1+1]=temp
    return mylist

def bubble_dsc(mylist):
    for idx in range(len(mylist)-1):
        for idx1 in range(len(mylist)-idx-1):
            if mylist[idx1]<mylist[idx1+1]:
                temp=mylist[idx1]
                mylist[idx1]=mylist[idx1+1]
                mylist[idx1+1]=temp
    return mylist
```

```python
#Task 9 


points = [(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]
distance = []
dict1 = {}

for i in points:
    dist = (i[0]**2 + i[1]**2)**(1/2)
    distance.append(dist)
    dict1[dist] = i

for index1 in range(0, len(distance)-1):
    mini_val = distance[index1]
    mini_index = index1
    for index2 in range(index1+1, len(distance)):
        if distance[index2] < mini_val:
            mini_val = distance[index2]
            mini_index = index2
    temp = mini_val
    distance[mini_index] = distance[index1]
    distance[index1] = temp
minimum = distance[0]
print("Minimum distance:",minimum)
print(f"Here the closest point is {dict1[minimum]} which has a distance of {minimum} from the origin.")
```
