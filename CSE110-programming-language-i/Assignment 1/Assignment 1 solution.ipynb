```python
#Task 14

distance=float(input('Input the distance in meters: '))
time=float(input('Input the time in seconds:'))

if (distance<0 or time<0):
    print('Time or distance cannot be negative')
else:
    velocity=(distance*3600)/(time*1000)
    print("Velocity:", velocity,"km/h")
    if velocity<60:
        print('Too slow. Needs more changes.')
    elif velocity>=60 and velocity<=90:
        print('Velocity is okay.The car is ready!')
    else:
        print('Too fast. Only a few changes should suffice.')
```
