```python
#Task 10

s=input('Please, enter the string: ')
s1=''
s2=''
new_string=''

for i in range(len(s)):
    if s[i] != ",":
        s1+=s[i]
    else:
        s2=s[i+2:]
        break
        
if len(s1)>len(s2):
    for i in range(len(s2)):
        new_string+=s1[i]
        new_string+=s2[i]
    new_string+=s1[len(s2):]
else:
    for i in range(len(s1)):
        new_string+=s1[i]
        new_string+=s2[i]
    new_string+=s2[len(s1):]

print(new_string)
```

```python
#Task 10

var=input("Please, enter two strings separated by comma and a space: ")
s1=""
s2=""
nstring=""

for i in var:
    if i==",":
        break
    else:
        s1+=i

s2=var[len(s1)+2:len(var)+1]

if len(s1)<len(s2):
    for i in range(len(s1)):
        nstring+=(s1[i]+s2[i])
    nstring+=s2[len(s1):]
else:
    for i in range(len(s2)):
        nstring+=(s1[i]+s2[i])
    nstring+=s1[len(s2):]

print(nstring)
```
