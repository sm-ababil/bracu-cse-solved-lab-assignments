```python
#Tvarsk 9

var="7 12 4 55 91 2 11 61 33 42"
list1=[]
list2=[]

for i in range(len(var)):
    if " " not in var:
        list1.append(int(var))
        if int(var)%2 != 0:
            list2.append(int(var))
        break
    else:
        b=var[:var.find(" ")]
        list1.append(int(b))
        if int(b)%2 != 0:
            list2.append(int(b))
        var=var[var.find(" ")+1:]

print("Originvarl list:",list1)
print("Modified list:",list2)
```

```python
#Task 10

var=input("Please,input the list with a comma and a space between two numbers: ")
list1=var.split(", ")
mylist=[]

for i in range(len(list1)):
    mylist.append(int(list1[i]))
    
mod_list=[mylist[0]]
for i in range(1,len(mylist)):
    if mylist[i] not in mod_list:
        mod_list.append(mylist[i])
        
print("Input list:",mylist)
print("Modified list:",mod_list)
```
