```python
from sympy import*


x,y,z= symbols("x y z")
f = 3*x**4+ 4*x**3+5*x**2+6*x+7
df1 = diff(f, x)
df2 = diff(f, x, 2)
df3 = diff(f, x, 3)
print(df1)
print(df2)
print(df3)
f = lambdify(x, f)
df1 = lambdify(x, df1)
df2 = lambdify(x, df2)
df3 = lambdify(x, df3)
```

```python
import numpy as np

p = np.arange(-7,6)
a = f(p)
print(a)

q = np.arange(-5,6)
b = df1(q)
print(b)

r = np.arange(-8,8)
c = df2(r)
print(c)

s = np.arange(-9,2)
d = df3(s)
print(d)
```
