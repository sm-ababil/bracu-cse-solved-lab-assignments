```python
from sympy import*
init_printing()
x = symbols("x")
y = Function("y")
y1 = Derivative(y(x), x)
eqq = y1 + (y(x)**2)/x**2
sol = dsolve(eqq,y(x), ics = {y(1):"1/2"})
pprint(sol)
```

```python
from sympy.plotting import plot

p1 = plot(sol.rhs, (x, -1, 1/3-(1e-3)), ylim = [-10,10])
p2 = plot(sol.rhs, (x, 1/3+(1e-3),1), ylim = [-10,10])
p1.extend(p2)
p1.show()
```
